```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
base = pd.read_csv('dados/final.csv', index_col=0)
base.head()
```

```python
fig = plt.figure(figsize=(15,8))
fig.suptitle('Notas do ENEM por escola')
sns.set_palette('viridis')
sns.displot(data=base, x='nota enem', hue='tp_dependencia', col='tp_localizacao', kind='hist')

plt.show()
```
